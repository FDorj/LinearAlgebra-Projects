```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the dataset
data_set = pd.read_csv('./dataset/house_dataset.csv')
area = data_set['Area']
price = data_set['Price']
```

```python
# Calculate the slope (w) and y-intercept (b) using the normal equations
N = len(area)
sum_x = np.sum(area)
sum_y = np.sum(price)
sum_xy = np.sum(area * price)
sum_x_squared = np.sum(area ** 2)
w = (N * sum_xy - sum_x * sum_y) / (N * sum_x_squared - sum_x ** 2)
b = (sum_y - w * sum_x) / N
```

```python
# Predicted price based on the model
predicted_price = w * area + b
```

```python
# Calculation of Mean Squared Error
mse = np.mean((price - predicted_price) ** 2)
```

```python
# Define the predict function
def predict(x):
    return w * x + b
```

```python
# Plotting the results
plt.scatter(area, price, color='blue', label='Data points')
plt.plot(area, w * area + b, color='red', label='Regression line')
plt.xlabel('Area (square meters)')
plt.ylabel('Price (currency units)')
plt.title('House Price Prediction with Actual Dataset')
plt.legend()
plt.show()
```

```python
# Print model parameters and MSE
print(f"Slope (w): {w}")
print(f"Y-intercept (b): {b}")
print(f"Mean Squared Error (MSE): {mse}")
```

```python
# Test cases
test_cases = [70, 110, 190, 260, 300, 370]
for tc in test_cases:
    print(f"Predicted price for area {tc}: {predict(tc)}")
```
